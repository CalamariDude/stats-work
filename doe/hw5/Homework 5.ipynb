# 1

```r
data <- read.table("./CH19PR14.txt",header=FALSE)
data
```

```r
y <- data$V1
A <- data$V2
B <- data$V3
```

## 2

```r
A <- factor(data$V2, 
                  levels = c(1, 2, 3),
                  labels = c("1", "2", "3"))
B <- factor(data$V3,
                   levels=c(1,2,3),
                   labels=c("1","2","3"))

#plot the amount of each treatment
table(A, B)
```

```r
model <- aov(V1 ~ V2 * V3, data = data)
```

```r
fit <- lm(V1 ~ V2 * V3, data=data)
```

## 3

```r
plot(model)
```

Focusing only on the residual plot, it can be noted that there appears to be some outliers in the data, which can suggest some nonlinearity. There appears to be bouncing around which indicates the residuals are ill-behaved. There is a general horizontal trend to the residuals, so this indicates that the variance of the error terms in approximitely equal.

## 4

As you can see above, the normal QQ plot seems to have a light tail, but inhabits a "gap" which can indacate some bimodal distriubtion (although not enough evidence to conclude this). Overal, the normality assumption still apears reasonable.

## 5

```r

plot.design(y ~ A*B, data = data )
```

Yes, the graph shows that there is some factor level difference. It is hard to tell without taking into account the variance. But there is defintely some factor effects when looking at the means.

## 6

```r
interaction.plot(A, B, y)
```

There appears to be some interaction between A and B as the main effect lines are not totally parallel. Especially when A changes from 2 to 3.

## 8

The two factors interact, as the P value for their terms is < .05, this indicates that we can reject that hypothesis that they don't interact and conclude that they do. We should NOT drop the interaction term from our model.
