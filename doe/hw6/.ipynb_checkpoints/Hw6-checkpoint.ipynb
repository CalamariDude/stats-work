## Part I

```r
data <- read.table("./APC1.txt",header=FALSE)
summary(data)
```

```r
model <- aov(V2 ~ V9, data = data)
```

```r
plot(model)
```

### 1
In the standardized residuals vs fitted values plot, the error variance seems to not have the same spread in the four geographic regions. This allows us to assert that the error variances are not approximately equal for the four regions.

### 2

```r
#mean
tapply(data$V2, data$V9, mean)
```

```r
#normal
tapply(data$V2, data$V9, sd)
```

```r
#sqrt
tapply(sqrt(data$V2), data$V9, sd)
```

```r
#log
tapply(log(data$V2), data$V9, sd)
```

```r
#reciprocal
tapply(1/data$V2, data$V9, sd)
```

It seems as though the reciprocal transformation yields error variances that are the closest to each other.

### 3

```r
rec_aov = aov(1/V2 ~ V9, data=data)
plot(rec_aov)
```

The standardized residuals against fitted values plot shows that the error variance is now more approximately equal amongst the 4 groups. In additin, the normal probability plot of the residuals shows an approximately normal distribution of residuals, as there seems to be no major skewing, nor heavy outliers.

### 4

```r
summary(rec_aov)
```

Hypothesis:
$$ H_0 : \mu_0 = \mu_1 = \mu_2 = \mu_3 $$
<br>
$$ H_a : \text{at least one } \mu_i \neq \mu_j \text{ for }i\neq j\text{ and }i=0,1,2,3 \text{, } j=0,1,2,3$$
<br>
Based on the one-way ANOVA yielind p-value = 1.78e-9, we reject the null hypothesis and conclude $H_a$ with a significance level of .01. This means there is at least on difference in the mean length of stay for the geographic regions.

### 5 
We determined that there is a least one difference of means, but we don't know which one. So we will carry out a pairwise post hoc test to see which pairs are statistically different and how much

```r
pairwise.t.test(1/data$V2, data$V9, p.adj = "none")
```

We find that there is a statistically significant different with significance value $\alpha=$.01 between all pairs of geographic regions EXCEPT between 2 and 3, which have a p-value $.1184 > .01$. Therefore, all geographic regions BUT regions 2 and 3 have a difference of mean length of stay that is statistically significant.

## Part II

### 1

```r
data2 <-data[data$V1 %in% c(1:44,46,48,51,53,57,58,60,63,66,74,76,79,80,83,84,88,94,101,103,111),]
```

### 2

```r
data2$V3 = cut(data2$V3, breaks=c(0,54,120), right=FALSE, labels=c("<54", "54+"))
aov2 <- aov(V2 ~ V9 * V3, data = data2)
```

### 3

```r
plot(aov2)
```

The residuals vs fitted plot looks to have approximatly equal error variance. We shall therefore assume constancy of error variance for our analysis.
The Normal Probability Plot of the residuals (QQ) seems normal as well, meaning that the residuals are normally distributed. We can now proceed with our ANOVA test

### 4

```r
plot.design(V2 ~ V3 + V9, data=data2)
```

```r
interaction.plot(data2$V3, data2$V9, data2$V2)
```

Based on these plots, it is apparent that the there seems to be a difference in the factor effects between geographic distance, but not so much with age. We will conduct ANOVA to find out. In addition, it appears that there is some interaction between the age and the first geographic region.

### 5

```r
summary(aov2)
```

Based on a significance level of $\alpha=.05$, it appears that there are no interaction effects present between age and geographic region since Pvalue for age*region is .08 > .05.

### 6

Based on the P-value for age, .5177, which is more than .05, there appears to be no main effects present.
Based on the P-value for region, 1e-7, which is less than .05, there appears to be a main effect present.
